# Classification

```python
from matplotlib import pylab as plt
import numpy as np
import pandas as pd
from pandas import DataFrame

import sklearn
sklearn.__version__
```

## Load some example data

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

df = DataFrame(
    data = iris.data, 
    columns = iris.feature_names
)

df['class'] = iris.target_names[iris.target]
df['target'] = iris.target

df.loc[0:150:10]
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, c=df.target, figsize = (12,12));
```

## Binary Classification

```python
X = df[['sepal width (cm)', 'petal length (cm)']]
```

### Make binary target variable

```python
Y_bin = (df.target == 0)
```

```python
scatter_matrix(X, c=Y_bin, figsize=(10,10), marker='.', s=70);
```

## Use Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(solver='lbfgs')
classifier.fit(X, Y_bin);
```

BFGS: Broyden-Fletch-Goldfarb-Shanno alogirthm

```python
from sklearn.metrics import accuracy_score

accuracy_score (Y_bin, classifier.predict(X))
```
